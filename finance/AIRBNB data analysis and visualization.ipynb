## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
df.isnull().sum().sort_values()
```

```python
# Find categorical columns in the DataFrame
categorical_columns = df.select_dtypes(include=['object']).columns
# Perform imputation for each categorical column
for column in categorical_columns:
 most_frequent_category = df[column].mode()[0]
 df[column].fillna(most_frequent_category, inplace=True)
```

```python
# Find numerical columns in the DataFrame
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
# Perform imputation for each numerical column
for column in numerical_columns:
 mean_value = df[column].mean()
 df[column].fillna(mean_value, inplace=True)
```

## Task 5b: Data Visualization (Any Tool)

* Does service price and room price have an impact on each other. Illustrate this relationship with a scatter plot and state your inferences
* Using a line graph show in which year the maximum construction of rooms took place.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# Extract the data for the scatter plot
service_fee = df['service_fee']
room_price = df['price']
# Create the scatter plot
plt.figure(figsize=(8, 8)) 
plt.scatter(service_fee, room_price, alpha=0.6)
plt.xlabel('service_fee')
plt.ylabel('Room Price')
plt.title('Scatter Plot: service_Fee vs. Room Price')
plt.show()
```

Insight: there is positiv relationship between service _fee and Room_price 

```python
# Group the data by year and calculate the sum of calculated_host_listings_count for each year
listings_per_year = df.groupby('construction_year')['calculated_host_listings_count'].sum()
# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(listings_per_year.index, listings_per_year.values, marker='o', linestyle='-', color='b')
plt.xlabel('Construction Year')
plt.ylabel('Total Calculated Host Listings')
plt.title('Total Calculated Host Listings Per Construction Year')
plt.grid(True)
plt.show()
```

Insight:The most constuction year took place between 2017 to nearly 2019
