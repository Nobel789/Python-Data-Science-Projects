```python
#adding diagonal
# we are taking g, using dot axes to acess all of these element of facedgrid
#and use the flat to flaten out this area , so that we can iterate about it
g= sns.FacetGrid(df,row='Genre',col='releaseyear',hue='Genre')
kws = dict(s=75, linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter,'CriticRatings', 'AudienceRatings',**kws)
for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c='gray',ls='--')
g.add_legend()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('dark', {'axes.facecolor': 'black'})

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Plot KDEs
f1 = sns.kdeplot(x=df['BudgetMillion'], y=df['AudienceRatings'], shade=True, shade_lowest=True, cmap='inferno', ax=axes[0, 0])
f1b = sns.kdeplot(x=df['BudgetMillion'], y=df['AudienceRatings'], cmap='cool', ax=axes[0, 0])
f2 = sns.kdeplot(x=df['BudgetMillion'], y=df['CriticRatings'], shade=True, shade_lowest=True, cmap='inferno', ax=axes[0, 1])
f2b = sns.kdeplot(x=df['BudgetMillion'], y=df['CriticRatings'], cmap='cool', ax=axes[0, 1])

# Plot lmplot
f3 = sns.lmplot(data=df, x='CriticRatings', y='AudienceRatings', fit_reg=False, hue='Genre', height=6, aspect=1.5)

# Plot Violin plots
f4 = sns.violinplot(data=df[df.Genre == 'Action'], x='releaseyear', y='AudienceRatings', ax=axes[1, 1], palette='Set2')
f5 = sns.violinplot(data=df[df.Genre == 'Drama'], x='releaseyear', y='AudienceRatings', ax=axes[2, 0], palette='Set2')
f6 = sns.violinplot(data=df[df.Genre == 'Comedy'], x='releaseyear', y='AudienceRatings', ax=axes[2, 1], palette='Set2')

# Add legend to the last three plots
f4.legend(title='Genre', labels=['Action'])
f5.legend(title='Genre', labels=['Drama'])
f6.legend(title='Genre', labels=['Comedy'])

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
f1=sns.kdeplot(x=df['BudgetMillion'], y=df['AudienceRatings'],shade=True,cmap='Reds',ax=axes[0,0])
f2=sns.kdeplot(x=df['BudgetMillion'], y=df['CriticRatings'],shade=True,cmap='Reds',ax=axes[0,1])
f3=sns.kdeplot(x=df['CriticRatings'], y=df['AudienceRatings'],shade=True,shade_lowest=False,cmap='Reds',ax=axes[1,0])
f4=sns.violinplot(data=df[df.Genre=='Action'], x='releaseyear',y='AudienceRatings',ax=axes[1,1])
f5=sns.violinplot(data=df[df.Genre=='Drama'], x='releaseyear',y='AudienceRatings',ax=axes[2,0])
f6=sns.violinplot(data=df[df.Genre=='Comedy'], x='releaseyear',y='AudienceRatings',ax=axes[2,1])
# Add legend to the last three plots
f4.legend(title='Genre', labels=['Action'])
f5.legend(title='Genre', labels=['Drama'])
f6.legend(title='Genre', labels=['Comedy'])

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```
